```python
import math
def f(x):
    x=float(x)
    return math.sin(x/5)*math.exp(x/10) + 5*math.exp(-x/2)
A=[]
b=[]
A.append([1, 1, 1, 1])
b.append(f(1))
A.append([1, 4, 16, 64])
b.append(f(4))
A.append([1, 10, 100, 10**3])
b.append(f(10))
A.append([1, 15, 225, 15**3])
b.append(f(15))
from scipy import linalg
x = linalg.solve(A, b)
print x
from matplotlib import pylab as plt
y1,y2=[],[]
for i in range(1,16):
    y1.append(f(i))
    y2.append(x[0]+i*x[1]+i**2*x[2]+i**3*x[3])
plt.plot(y1)
plt.plot(y2)
sub = open('submission-2.txt', 'w')
sub.write(x.__str__())
sub.close()
```
